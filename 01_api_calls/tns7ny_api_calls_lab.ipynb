## **Q2**

```python
# 2 - Accessing saferproducts.gov API in a notebook

import urllib.request
import json
import pandas as pd

url = 'https://www.saferproducts.gov/RestWebServices/' # Location of the API
query = 'Recall?format=json&ProductType=Exercise' # The query
response = urllib.request.urlopen(url+query)
response_bytes = response.read()
data = json.loads(response_bytes) # Convert response to json
response.close()

df = pd.DataFrame.from_dict(data)
df.head()
```

```python
# 3 - Convert code chunk into a function you can reuse to flatten dictionaries,
# or explain problems you run into while attempting to do so. Make some tables or plots


# Original code chunk
temp = df['RemedyOptions']
clean_values = []
for i in range(len(temp)):
    if len(temp[i])>0:
        values = []
        for j in range(len(temp[i])):
            values.append(temp[i][j]['Option'] )
        clean_values.append(values)
    else:
        clean_values.append('')
df['remedy'] = clean_values

# --------------------------------------------

# Function
def flatten_dict(df, column_name):
    clean_values = []
    for i in range(len(df[column_name])):
        if len(df[column_name][i]) > 0:
            values = []
            for j in range(len(df[column_name][i])):
                values.append(df[column_name][i][j]['Option'])
            clean_values.append(values)
        else:
            clean_values.append('')
    df[column_name + '_clean'] = clean_values
    return df


# Example table
data = {
    'ID': [1, 2, 3, 4],
    'RemedyOptions': [
        [{'Option': 'Refund'}, {'Option': 'Exchange'}],
        [{'Option': 'Repair'}, {'Option': 'Replace'}],
        [],
        [{'Option': 'Store Credit'}, {'Option': 'Gift Card'}]
    ]
}

df = pd.DataFrame(data)

df = flatten_dict(df, 'RemedyOptions')

print(df[['ID', 'RemedyOptions', 'RemedyOptions_clean']])
```
